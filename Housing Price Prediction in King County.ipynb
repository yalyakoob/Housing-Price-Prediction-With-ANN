# Data Preprocessing For Model

```python
X = bottom_99.drop('price',axis=1).values #remove target variable from training data
y = bottom_99['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
#Data scaling 
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Building and Training our Model 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
#adding 4 hidden layers with 17 neurons for each layer 
model.add(Dense(17,activation='relu'))
model.add(Dense(17,activation='relu'))
model.add(Dense(17,activation='relu'))
model.add(Dense(17,activation='relu'))
#output layer 
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train, validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
```

# Model Evaluation

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Based on the plot above we can see that there are no signs of overfitting as the training loss is consistent with validation and does not drop below the validation loss at the different epochs of training. We can potentially continue training the model as the validation loss does not seem to have reached an inflection point in which it is increasing

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

Based on the mean_absolute_error we can see that our predictions are off by roughly 95000 dollars

Another metric we can can evaluate is the explained variance score. Since a primary objective of any regression problem is to explain the variability in the dependent variable given the variability in the dependent variable we can use the explained variance score to see how much variability in the dependent variable our model explains

```python
explained_variance_score(y_test,predictions) #model explains about 74 % of variability in dependent variable 
```
